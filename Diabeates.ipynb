```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
X, y = datasets.load_diabetes(return_X_y=True)

# Print the shape of the data snd the first row
print(X.shape)
print(X[0])
```

```python
# Extract the column at Index 2
X = X [:, 2]
print(X.shape)

# Reshape to a 2D array
X = X.reshape((-1,1))
print(X.shape)
print(X)
```

```python
# split the data into training and testing data
from sklearn import model_selection
X_train,X_test, y_train, y_test = model_selection.train_test_split(X,y, test_size = 0.33)
```

```python
# Create  a linear regression model and train it with our data
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict using our test data
y_pred = model.predict(X_test)
```

```python
# Create a scatter plot
plt.scatter(X_test,y_test, color ='black')

# Plot the prediction
plt.plot(X_test,y_pred, color = 'blue', linewidth = 3)

# Add labels and title

plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabeates Progression against BMI')

# Draw the plot
plt.show()
```
